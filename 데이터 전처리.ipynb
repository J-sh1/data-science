### 2012년 데이터

```python
data1.columns
```

```python
data1
```

```python
data1 = data1[data1['2012학년도 대학 입학전형 수시모집 결과분석 자료'] == '광주']
res_data1 = data1[['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 10', 'Unnamed: 11']]
res_data1 = res_data1.rename(columns = {
    'Unnamed: 2' : '전형구분명',
    'Unnamed: 3' : '대학명',
    'Unnamed: 4' : '학과명',
    'Unnamed: 5' : '모집인원',
    'Unnamed: 6' : '지원인원',
    'Unnamed: 8' : '경쟁률',
    'Unnamed: 10' : '평균',
    'Unnamed: 11' : '표준편차'
})
res_data1['연도'] = 2012
res_data1
```

```python
res_data1.info()
```

```python
res_data1 = res_data1.dropna(axis=0)
```

```python
res_data1['모집인원'] = pd.to_numeric(res_data1['모집인원'], errors='coerce').astype('Int64')
res_data1['지원인원'] = pd.to_numeric(res_data1['지원인원'], errors='coerce').astype('Int64')
res_data1['경쟁률'] = pd.to_numeric(res_data1['경쟁률'], errors='coerce')
res_data1['평균'] = pd.to_numeric(res_data1['평균'], errors='coerce')
res_data1['표준편차'] = pd.to_numeric(res_data1['표준편차'], errors='coerce')

res_data1.info()
```
