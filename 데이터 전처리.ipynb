### 2013년 데이터

```python
data2.columns
```

```python
data2
```

```python
# 사용한 컬럼만 필터링 후 저장
data2 = data2[data2['2013학년도 대학 입학전형 수시모집 결과분석 자료'] == '광주캠퍼스']
res_data2 = data2[['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 8','Unnamed: 10', 'Unnamed: 11']]
res_data2 = res_data2.rename(columns = {
    'Unnamed: 2' : '전형구분명',
    'Unnamed: 3' : '대학명',
    'Unnamed: 4' : '학과명',
    'Unnamed: 5' : '모집인원',
    'Unnamed: 6' : '지원인원',
    'Unnamed: 8' : '경쟁률',
    'Unnamed: 10' : '평균',
    'Unnamed: 11' : '표준편차'
})
res_data2['연도'] = 2013
res_data2
```

```python
# 결측치 확인
res_data2.info()
```

```python
# 결측치 있는 행 제거
res_data2 = res_data2.dropna(axis=0)
```

```python
# 모집인원, 지원인원은 정수형으로 변환
res_data2['모집인원'] = pd.to_numeric(res_data2['모집인원'], errors='coerce').astype('Int64')
res_data2['지원인원'] = pd.to_numeric(res_data2['지원인원'], errors='coerce').astype('Int64')

# 경쟁률, 평균, 표준편차는 부동소수점형으로 변환
res_data2['경쟁률'] = pd.to_numeric(res_data2['경쟁률'], errors='coerce')
res_data2['평균'] = pd.to_numeric(res_data2['평균'], errors='coerce')
res_data2['표준편차'] = pd.to_numeric(res_data2['표준편차'], errors='coerce')

res_data2.info()
```
